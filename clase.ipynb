```python
import pandas as pd
netflix = pd.read_excel('netflix.xlsx')
#?eliminacion de datos
#* sino pongo inplace= no se borra
print(f"los numeros son: {netflix.shape}")#?muestra fila y columnas
print(netflix.drop(columns=['listed_in', 'cast'], inplace=True))#?elimina columnas
print(netflix.shape)

#?eliminar filas
print(netflix.drop(range(50,2500), inplace=True))#? elimina en un rango, y drop elimina filas for defecto para columnas pones columns
print(netflix.shape)
print(netflix.drop([1, 10, 5], inplace=True)) #?eliminas las filas que quieres
print(netflix.shape)

#? eliminación condicional
print(netflix.shape)
print(netflix.drop(netflix[netflix['type'] == "Tv show"].index)) #?para codincional siempre poner index para obviar los indices, y que type columna sea simila tv show
```

```python

#* manejo de datos rapidos
#* si quiero manejar datos vacios, pongo None o importo la libreria numpy: np.nan
import pandas as pd
import numpy as np
data={
    'A':[1, 2, None, 4],
    'B':[0, None, 3, 4],
    'C': [np.nan, None, "c",5]
}

df = pd.DataFrame(data)
print(df)
print(df.isnull())#? muestra si es nulo
print(df.isna())#?muestra sin info, vacio
print(df.notnull())#? muestra si es nulo
print(df.notna())#?muestra sin info, vacio

print(df.dropna())#?elimina filas con valores vacios
print(df.dropna(axis=1)) #?elimina columna con valores vacios
print(df.dropna(thresh=2)) #?elimina las filas con almenos 2 valores nulos
```

```python
#metodo sencillo para limpiar información
print(df.fillna("hola")) #?rellena los elementos vacios con str
print(df.fillna(df["B"].mean()))#? solo a la columna B rellenalo y saca un promedio
print(df.fillna(method='ffill'))#?rellena con el numerio anterior
print(df.fillna(method=('bfill')))#? rellena con los elemntos siguientes
```
